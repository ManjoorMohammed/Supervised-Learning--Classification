```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris= load_iris()
iris
```

```python
X = iris.data  # Features
y = iris.target  # Labels

print (X)
print (y)
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)
```

## Logical Regression 

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Initialize the logistic regression model
model = LogisticRegression()
model
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
y_pred
```

```python
# Calculate accuracy
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Example: Predicting new samples
new_samples = [[5.1, 3.5, 1.4, 0.2],  # Example Setosa
               [6.0, 3.0, 4.5, 1.5],  # Example Versicolour
               [6.7, 3.1, 5.6, 2.4]]  # Example Virginica

# Use the trained model to predict the labels of new samples
predictions = model.predict(new_samples)

# Map predicted labels to actual class names
predicted_classes = [iris.target_names[pred] for pred in predictions]

# Print the predictions
print("\nPredictions for new samples:")
for i in range(len(new_samples)):
    print(f"Sample {i+1}: {predicted_classes[i]}")
```

```python
from sklearn.metrics import confusion_matrix
# Print confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)

# Plot confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## KNN

## KMeans

```python
from sklearn.cluster import KMeans
```

```python
# Load Iris dataset
X = iris.data[:, :4]  # Specific features: sepal_length, sepal_width, petal_length, petal_width

# Applying K-Means with K=3 (since we know there are 3 classes in the Iris dataset)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
# Getting cluster centers and labels
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Visualizing the clusters (considering only two features for visualization)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200, marker='o', label='Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clustering on Iris Dataset')
plt.legend()
plt.show()
```
